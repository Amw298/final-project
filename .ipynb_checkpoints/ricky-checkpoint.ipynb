SubMetric: Housing price and value. Description: Shows the median price of housing in each neighborhood to see what the cheapest neighborhoods are for purchasing a house. This is a metric that will greatly impact how accesible it is for a person to purchase their own property.

```python
import pandas as pd
hv = pd.read_csv("Housing.csv")
hv = hv.sort_values('Median Home Value')
hve =hv.iloc[:,[0,21]]
hve = hve.dropna()
hve
```

```python
hve.plot.bar(x='Neighborhood',figsize=(15,3))
```
